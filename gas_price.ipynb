```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Line Graph

```python
gas = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/matplotlib/gas_prices.csv')
 
plt.figure(figsize=(11,5), dpi=135)
 
plt.title('Gas Prices over Time (in USD)')
 
#plt.plot(gas.Year, gas.USA, '*-', label='United States')
#plt.plot(gas.Year, gas.Canada, '*-', label='Canada')
#plt.plot(gas.Year, gas['South Korea'], '*-', label='South Korea')
 
label = np.array(gas.columns.values[1:11])
n = 0
 
for country in gas:
   if country != 'Year':
      plt.plot(gas.Year, gas[country], '*-', label=label[n])
      n+=1
          
thrd_yr = gas.Year[::3]
 
plt.xticks(thrd_yr)
plt.xlabel('Year')
plt.ylabel('Gas Prices')
 
plt.legend()
 
plt.show()
```
