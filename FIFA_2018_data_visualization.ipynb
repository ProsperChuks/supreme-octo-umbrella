```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

text_kwargs = dict(fontsize=24)
fifa = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/matplotlib/fifa_data.csv')
fifa.head(5)
```

### **Histogram**

```python
bin = np.array([40,50,60,70,80,90,100])

plt.figure(figsize=(10,5))

plt.hist(fifa.Overall, bins=bin, color='red')

plt.xticks(bin)
plt.ylabel('Number of Players')
plt.xlabel('Skill level')
plt.title('Distribution of Player Skills in FIFA 19', **text_kwargs)

plt.savefig('/content/drive/MyDrive/Colab Notebooks/matplotlib/fifa_overall hist.png')

plt.show()
```
